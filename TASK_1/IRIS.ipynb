# MODEL TRAINING

## Splitting Data

```python
from sklearn.model_selection import train_test_split

features = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']

#defining the feature matrix
x = df.loc[:, features].values
y = df.Species

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 40,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

## Metrics for Evaluating Models

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import  classification_report
from sklearn.metrics import confusion_matrix
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg_model= LogisticRegression(random_state = 0)
logreg_model.fit(x_train, y_train)

# Model training
logreg_model.fit(x_train, y_train)

# Prediction
prediction=logreg_model.predict(x_test)
```

### Prediction

```python
print('Prediction:\n',prediction)
```

### Accuracy

```python
print('Accuracy:',metrics.accuracy_score(prediction,y_test))
```

### Classification Report

```python
print('Classification Report:\n',classification_report(y_test,prediction))
```

### Confusion Matrix

```python
con_mat=confusion_matrix(y_test,prediction)
label = [0,1,2]

plt.figure(figsize =(6,4))
sns.heatmap(con_mat, annot=True, fmt='d',cmap='RdPu',cbar=False,xticklabels=label, yticklabels =label)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
print('Confusion Matrix:\n',confusion_matrix(y_test,prediction ))
```

# Support Vector Machine

```python
from sklearn.svm import SVC
svm_model=SVC()

# Model training
svm_model.fit(x_train, y_train)

# Predicting
svm_prediction=svm_model.predict(x_test)
```

### Prediction
